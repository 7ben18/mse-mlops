# Melanoma Dataset EDA

This notebook reads the dataset location from `config/train.yaml` and inspects the raw `train` and `test` image folders used by the project.

```python
from IPython.display import display

from mse_mlops.analysis.melanoma_dataset import (
    build_class_distribution_frame,
    build_overview_frame,
    load_dataset_overview,
    make_contact_sheet,
    select_class_records,
)
```

```python
overview = load_dataset_overview()

print(f"Config path: {overview.config_path}")
print(f"Dataset root: {overview.data_dir}")
print(f"Train dir: {overview.train_dir}")
print(f"{overview.eval_split_name.capitalize()} dir: {overview.eval_dir}")
print(f"Validation mode in config: {overview.validation_mode}")
print(f"Classes found: {overview.classes}")
print(f"Total samples: {len(overview.all_records)}")
print(f"Number of classes: {len(overview.classes)}")
print("Classes: " + ", ".join(overview.classes))
```

```python
display(build_overview_frame(overview))
```

```python
display(build_class_distribution_frame(overview))
```

```python
print("Example images per class")
samples_per_class = 8

for class_name in overview.classes:
    selected_records = select_class_records(overview, class_name, sample_size=samples_per_class)
    print(class_name)
    display(make_contact_sheet(selected_records, columns=samples_per_class))
```
